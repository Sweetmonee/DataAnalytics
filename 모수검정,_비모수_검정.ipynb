두개 학급의 시험성적에 대한 데이터이다. 두 학습의 시험 평균(비모수검정의 경우 중위값)은 동일하다 말할 수 있는지 확인 하라

```python
import pandas as pd

df1 = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/scipy/ind1.csv')
df2 = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/scipy/ind2.csv')


# 표본수가 100개 이므로 샤피로 정규성 검정 수행(5000개 이하)

from scipy.stats import shapiro

shapiro_stats_df1, p_value_df1 = shapiro(df1)
shapiro_stats_df2, p_value_df2 = shapiro(df2)
print(shapiro_stats_df1, p_value_df1)
print(shapiro_stats_df2, p_value_df2)
# p_value 값이 0.05보다 크므로 귀무가설 채택, 정규성 만족

# 등분산성 검정 시행

from scipy.stats import bartlett
from scipy.stats import levene
from scipy.stats import fligner

bartlett_stats, bartlett_p_value = bartlett(df1['data'], df2['data'])
levene_stats, levene_p_value = levene(df1['data'], df2['data'])
fligner_stats, fligner_p_value = bartlett(df1['data'], df2['data'])

print(bartlett_stats, bartlett_p_value)
print(levene_stats, levene_p_value)
print(fligner_stats, fligner_p_value)

# 유의수준 0.05에서 levene만 등분산성을 만족한다고 하지만... 데이터마님에서는 levene만 사용하였으므로 나도 똑같이 해보겠다. 귀무가설을 채택한다.

from scipy.stats import ttest_ind

ttest_ind_stats, ttest_ind_p_value = ttest_ind(df1['data'], df2['data'], equal_var = True)
print(ttest_ind_stats, ttest_ind_p_value)

# 측정결과 p_value가 유의 수준인 0.05보다 낮게 나왔으므로 귀무가설을 기각한다.(alternative = 'Two_sided'가 디폴트)

print(f't_test 독립표본 검정결과 p_value는 {ttest_ind_p_value}가 나왔습니다. 귀무가설을 기각합니다. 두 표본의 평균은 같지 않습니다.')
```

```python
두개 학급의 시험성적에 대한 데이터이다. 두 학습의 시험 평균(비모수검정의 경우 중위값)은 동일하다 말할 수 있는지 확인 하라
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/scipy/scipy5.csv')

A = df['A']
B = df['B'].dropna()
# 정규성 검정
from scipy.stats import shapiro

shapiro_stats_A, shapiro_p_values_A = shapiro(A)
shapiro_stats_B, shapiro_p_values_B = shapiro(B)

print(shapiro_stats_A, shapiro_p_values_A)
print(shapiro_stats_B, shapiro_p_values_B)

# 결과 두 표본 다 정규성을 만족시키지 못한다. 비모수 검정을 시행한다.

from scipy.stats import mannwhitneyu
from scipy.stats import ranksums

mannwhitneyu_stats_A, mannwhitneyu_p_value_A = mannwhitneyu(A, B)
ranksums_stats_A, ranksums_p_value_A = ranksums(A, B)

print(mannwhitneyu_stats_A, mannwhitneyu_p_value_A)
print(ranksums_stats_A, ranksums_p_value_A)

# 둘 다 p_value 값이 0.05를 넘었으므로 중위값은 동일하다고 할 수 있다.
```

```python
두개 그룹에 대한 수치형 데이터이다. 두 그룹의 평균은 동일하다 말할 수 있는지 검정하라
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/scipy/ind3.csv')

print(df)

a = df[df['group'] == 'a']['data']
b = df[df['group'] == 'b']['data']

# 표본이 5000개 보다 작으므로 shapiro 정규성 검정 실시

from scipy.stats import shapiro

shapiro_stats_a , shapiro_p_value_a = shapiro(a)
shapiro_stats_b , shapiro_p_value_b = shapiro(b)

print(shapiro_stats_a , shapiro_p_value_a)
print(shapiro_stats_b , shapiro_p_value_b)

# 둘다 p_value 값이 0.05 이상이므로 귀무가설 채택, 정규성을 만족한다.

# ttest_ind 사용하기 전 등분산성 만족하는지 확인

from scipy.stats import levene
from scipy.stats import fligner
from scipy.stats import bartlett

levene_stats, levene_p_value = levene(a,b)
fligner_stats, fligner_p_value = fligner(a,b)
bartlett_stats, bartlett_p_value = bartlett(a,b)

print(levene_stats, levene_p_value)
print(fligner_stats, fligner_p_value)
print(bartlett_stats, bartlett_p_value)

# 등분산성 검정 결과 모두 유의수준인 0.05 이하로 p_value가 산출되었으므로, 귀무가설을 기각. 등분산성을 만족하지 않음

from scipy.stats import ttest_ind

ttest_ind_stats, ttest_ind_p_value = ttest_ind(a,b, equal_var = False)

print(ttest_ind_stats, ttest_ind_p_value)

#t_test 독립표본 검정결과 p_value가 유의수준보다 낮게 나왔으므로 귀무가설을 기각. 두 독립 표본의 평균은 같다고 볼 수 없음
```

두개 그룹에 대한 수치형 데이터이다. 두 그룹의 평균은 동일하다 말할 수 있는지 검정하라

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/scipy/ind6.csv')

# 널값 제거
a = df['a']
b = df['b'].dropna()

# 정규성 검정

from scipy.stats import shapiro

shapiro_stats_a, shapiro_p_value_a = shapiro(a)
shapiro_stats_b, shapiro_p_value_b = shapiro(b)

print(shapiro_stats_a, shapiro_p_value_a)
print(shapiro_stats_b, shapiro_p_value_b)

# 두 그룹다 정규성 만족

# 등분산성 검정

from scipy.stats import levene

levene_stats, levene_p_value = levene(a,b)

print(levene_stats, levene_p_value)
# 등분산성 만족하지 못한다.
# ttest_ind

from scipy.stats import ttest_ind

ttest_ind_stats, ttest_ind_p_value = ttest_ind(a,b, equal_var = False)

print(ttest_ind_stats, ttest_ind_p_value)

# 검정결과 유의수준 보다 p_value 값이 높게 나왔으므로 귀무가설을 채택. 두 대응 표본의 평균은 같다고 볼 수 있다.
```

특정 질병 집단의 투약 전후의 혈류량 변화를 나타낸 데이터이다. 투약 전후의 변화가 있는지 검정하라

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/scipy/rel2.csv')

#정규성 검정

before = df['before']
after = df['after']

from scipy.stats import shapiro # 5000개 이하이므로

shapiro_stats_before, shapiro_p_value_before = shapiro(before)
shapiro_stats_after, shapiro_p_value_after = shapiro(after)

print(shapiro_stats_before,shapiro_p_value_before)
print(shapiro_stats_after,shapiro_p_value_after)

#정규성 만족. ttest_rel 진행, 만약 만족 못했으면 wilcoxon 비모수검정 진행

from scipy.stats import ttest_rel

ttest_rel_stats, ttest_rel_p_value = ttest_rel(before, after, alternative = 'two-sided')

print(ttest_rel_stats, ttest_rel_p_value)

# 검정결과 P_value가 유의수준보다 낮으므로 귀무가설을 기각한다. 두 표본의 평균은 같다고 볼 수 없다.
```

특정 질병 집단의 투약 전후의 혈류량 변화를 나타낸 데이터이다. 투약 전후의 변화가 있는지 검정하라

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/scipy/rel3.csv')

# 정규성 검정

from scipy.stats import shapiro

before = df['before']
after = df['after']

shapiro_stats_before, shapiro_p_value_before = shapiro(before)
shapiro_stats_after, shapiro_p_value_after = shapiro(after)

print(shapiro_stats_before, shapiro_p_value_before)
print(shapiro_stats_after, shapiro_p_value_after)

# 정규성 만족
# 대응표본이므로 등분산성 검정 스킵

from scipy.stats import ttest_rel

ttest_rel_stats, ttest_rel_p_value = ttest_rel(before, after)

print(ttest_rel_stats, ttest_rel_p_value)

# 두 표본의 평균은 같다고 볼 수 있다. 귀무가설을 채택한다.
```

특정 집단의 학습 전후 시험 성적 변화를 나타낸 데이터이다. 시험 전과 후에 차이가 있는지 검정하라

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/scipy/rel1.csv')


before = df['before']
after = df['after']
# 정규성 검정

from scipy.stats import shapiro

shapiro_stats_before, shapiro_p_value_before = shapiro(before)
shapiro_stats_after, shapiro_p_value_after = shapiro(after)

print(shapiro_stats_before, shapiro_p_value_before)
print(shapiro_stats_after, shapiro_p_value_after)
# 검정결과 귀무가설 기각. 정규 분포를 만족시키지 않음
# 비모수 검정 진행

from scipy.stats import wilcoxon

wilcoxon_stats, wilcoxon_p_value = wilcoxon(before, after)

print(wilcoxon_stats, wilcoxon_p_value)
```

한 기계 부품의 rpm 수치를 두가지 다른 상황에서 측정했다.(총 70세트) b 상황이 a 상황보다 rpm값이 높다고 말할 수 있는지 검정하라

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/scipy/rel4.csv')

a = df[df['group'] == 'a']['rpm']
b = df[df['group'] == 'b']['rpm']

# 정규성 검정

from scipy.stats import shapiro

sha_stats_a, sha_p_value_a = shapiro(a)
sha_stats_b, sha_p_value_b = shapiro(b)
print(f'a의 shapiro 검정통계량은 {sha_stats_a}, p_value는 {sha_p_value_a} 입니다.')
print(f'b의 shapiro 검정통계량은 {sha_stats_b}, p_value는 {sha_p_value_b} 입니다.')
# 정규성을 만족한다.

from scipy.stats import ttest_rel

tr_stats, tr_p_value = ttest_rel(b, a, alternative = 'greater')

print(tr_stats, tr_p_value)

#귀무가설 기각. b 상황이 a 상황보다 rpm 값이 높다고 볼 수 있다.
```
