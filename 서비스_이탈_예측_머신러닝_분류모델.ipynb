```python
import pandas as pd
#데이터 로드
x_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churnk/X_train.csv")
y_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churnk/y_train.csv")
x_test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churnk/X_test.csv")

#EDA

# print(x_train.shape, y_train.shape, x_test.shape)
# print(x_train.isnull().sum(), y_train.isnull().sum(), x_test.isnull().sum())

# Data Preprocessing
# 연관성 없는 컬럼 제거
x_train = x_train.drop(['Surname'], axis = 1)
x_test = x_test.drop(['Surname'], axis = 1)
y_train_id = y_train.pop('CustomerId')
x_train_id = x_train.pop('CustomerId')
x_test_id = x_test.pop('CustomerId')

# feature scaling
from sklearn.preprocessing import LabelEncoder

object_columns = ['Geography','Gender']
le = LabelEncoder()
for oc in object_columns:
  x_train[oc] = le.fit_transform(x_train[[oc]])
  x_test[oc] = le.transform(x_test[oc])
# 정규화 미수행
# 피처스케일링 시행
# 표준화
# from sklearn.preprocessing import MinMaxScaler
# mms = MinMaxScaler()
# numeric_columns = ['CreditScore','Age','Tenure','Balance','EstimatedSalary']
# for nc in numeric_columns:
#   x_train[nc] = mms.fit_transform(x_train[[nc]])
#   x_test[nc] = mms.transform(x_test[[nc]])
# 정규화
# from sklearn.preprocessing import StandardScaler
# ss = StandardScaler()
# numeric_columns = ['CreditScore','Age','Tenure','Balance','EstimatedSalary']
# for nc in numeric_columns:
#   x_train[nc] = ss.fit_transform(x_train[[nc]])
#   x_test[nc] = ss.transform(x_test[[nc]])
print(x_train)

# train_test_split
from sklearn.model_selection import train_test_split

x_tr, y_tr, x_var, y_var = train_test_split(x_train, y_train, test_size = 0.2, random_state = 2023)
print()
# 모델링

from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators = 200, max_depth = 5, random_state = 2023)
rfc.fit(x_tr,x_var.values.ravel())
pred = rfc.predict(y_tr)
predict_proba = rfc.predict_proba(y_tr)[:,1]

# 평가

from sklearn.metrics import accuracy_score, f1_score, roc_auc_score

accracy_score_rfc_y_tr = accuracy_score(y_var, pred)
f1_score_rfc_y_tr = f1_score(y_var, pred, average = 'macro')
roc_auc_score_rfc_y_tr = roc_auc_score(y_var, predict_proba)
print(accracy_score_rfc_y_tr)
print(f1_score_rfc_y_tr)
print(roc_auc_score_rfc_y_tr)
# 수치형 변수 피처스케일링 안할시
# 0.8576923076923076
# 0.7215529587747866
# 0.8379268523360743

# 답안 제출
pred_submit = rfc.predict(x_test)
pred_proba_submit = rfc.predict_proba(x_test)[:,1]

df_submit = pd.DataFrame({'ID': x_test_id, 'Exist': pred_submit})
df_proba_submit = pd.DataFrame({'ID': x_test_id, 'Exist': pred_proba_submit})
df_submit.to_csv('submit.csv', index = False)
df_proba_submit.to_csv('submit_proba.csv', index = False)
```
